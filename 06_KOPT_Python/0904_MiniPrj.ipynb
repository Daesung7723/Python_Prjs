## 🚀 최종 실습 (업그레이드 버전): 상식 퀴즈 게임 만들기

#### 목표: 여러 주제의 퀴즈를 데이터로 만들어두고, 사용자에게 퀴즈를 출제하여 점수를 계산하고 결과를 보여주는 프로그램을 만듭니다.

#### 이 프로젝트에서 사용하는 핵심 개념들:

- 딕셔너리: 질문, 보기, 정답, 주제를 하나의 묶음(퀴즈 한 세트)으로 관리하는 데 아주 적합합니다.
- 리스트: 여러 개의 퀴즈 딕셔너리들을 순서대로 담아 전체 퀴즈 은행을 만드는 데 사용합니다.
- 집합(Set): 사용자가 맞힌 퀴즈의 '주제'들을 중복 없이 저장하여, 어떤 분야에 강한지 보여주는 데 사용합니다.
- 튜플: 각 퀴즈의 고유 번호와 질문 내용을 묶는 등, 절대로 바뀌면 안 되는 데이터를 안전하게 다룰 때 사용할 수 있습니다.
- 함수 (def): 전체 게임 로직을 play_quiz() 와 같은 하나의 함수로 깔끔하게 묶어 관리합니다.
- for 반복문: 퀴즈 은행에 있는 모든 퀴즈를 하나씩 출제하거나, 보기를 출력할 때 사용합니다.
- if/else 조건문: 사용자의 답변이 정답인지 아닌지 판별합니다.
- input() 함수: 사용자로부터 정답을 입력받습니다.

#### 게임 데이터 구조 설계:

1. 퀴즈 한 개 (딕셔너리):
    - "question": "한국의 수도는?" (질문)
    - "options": ["서울", "부산", "인천", "대구"] (보기 리스트)
    - "answer": "서울" (정답)
    - "topic": "상식" (주제)

2. 퀴즈 은행 (List): 위와 같은 딕셔너리 여러 개를 담고 있는 리스트를 만듭니다. (예: quiz_bank = [quiz1, quiz2, quiz3, ...])

#### 게임 진행 방식:
1. 게임이 시작되면, quiz_bank 리스트에 있는 퀴즈를 순서대로 하나씩 출제합니다.
2. 사용자에게 질문과 보기(1. 서울 2. 부산...)를 보여주고, 번호로 정답을 입력받습니다.
3. 사용자의 답변이 정답과 일치하는지 확인합니다.
4. 정답일 경우:
    - 점수를 10점 올려줍니다.
    - 해당 퀴즈의 주제("topic")를 '맞힌 주제' 집합(Set)에 추가합니다. (똑같은 주제를 또 맞아도 집합이라 중복 저장되지 않아요!)
5. 오답일 경우: "틀렸습니다! 정답은 OО이었습니다." 라고 알려줍니다.
6. 모든 퀴즈가 끝나면, 최종 점수와 함께 사용자가 맞힌 주제들의 목록(집합)을 출력하여 어떤 분야에 강한지 보여줍니다. (예: "당신은 [상식, 과학] 분야에 강하시군요!")
